```python
import numpy as np
import pandas as pd
```

# Load du lieu Year 1

```python
data = pd.read_csv('1.data-preparation/new-data/data_Y1.csv')
data.drop("MemberID", axis = 1, inplace = True)
data.drop("Year", axis = 1, inplace = True)

y = data['TARGET']
X = data.loc[:, data.columns != 'TARGET']
X.head()
```

```python
import math
def eval(model, X_test, y_test):
    pred_y = model.predict(X_test) # your predicted y values
    pred_y = pred_y.astype(int)
    # The root mean square logarithmic error
    msle = np.mean( (np.log(pred_y + 1) - np.log(y_test + 1)) ** 2)
    rmsle = math.sqrt(msle)
    return rmsle
```

# scale du lieu year 1

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler().fit(X)
new_X = scaler.transform(X)
```

# ap dung random forest len du lieu year 1

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(max_depth = 10, max_features = 50
                  , min_samples_leaf =  3, bootstrap = True, min_impurity_decrease = 0.0
                  , random_state = 3, n_estimators = 1000, n_jobs = -1, oob_score = False)
model.fit(new_X, y)
```

```python
eval(model, new_X, y)
```
